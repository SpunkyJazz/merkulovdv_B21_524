```python
import numpy as np
from PIL import Image as pim
```

```python
def bleaching(image):
  image_data = image.load()
  result_image = pim.new('L', image.size)
  new_image = result_image.load()
  for x in range(image.size[0]):
    for y in range(image.size[1]):
      new_image[x, y] = int(0.3 * image_data[x, y][0] + 0.59 * image_data[x, y][1] + 0.11 * image_data[x, y][2])
  return result_image
```

```python
input_image = pim.open("img1.png")
output_image = bleaching(input_image)
```
